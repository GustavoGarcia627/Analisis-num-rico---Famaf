**Ejercicio 3)** Obtener el mayor y menor numero positivo en punto flotante (overflow y underflow ). Para
obtener el mayor numero de overflow escribir un ciclo que vaya calculando las sucesivas
potencias de 2 y que finalice cuando se produce overflow. Se recomienda utilizar el comando
isinf (importar la librer ́ıa math) para detectar cuando se produce el overflow (escribir
help(math.isinf) para obtener informacion sobre este comando). Otra instruccion que
puede resultar util es break para interrumpir el ciclo cuando se produce el overflow, o
utilizar un while. El numero de underflow se puede obtener dividiendo por 2 repetidamente
hasta obtener un numero indistinguible del cero en punto flotante.

````python
import math

# Encontrar el mayor número antes del overflow
num = 1.0
while not math.isinf(num * 2):  # Multiplicar por 2 hasta que se vuelva infinito
    num *= 2

print("Mayor número antes del overflow:", num)

# Encontrar el menor número antes del underflow
num = 1.0
while num / 2 > 0:  # Dividir por 2 hasta que sea indistinguible de 0
    num /= 2

print("Menor número antes del underflow:", num)
````

**Ejercicio 8** Escribir dos funciones en Python para la resolucion de ecuaciones de segundo grado $ax^2 +
bx + c = 0.$ Una de ellas, que llamaremos mala, implementando la tradicional formula
de Baskhara y la otra, que llamaremos buena, usando una manera eficiente para evitar
cancelacion de dıgitos significativos. La sintaxis de la llamada a las funciones deber ́ıa ser:
`[x_1,x_2] = buena(a,b,c)`
Análogamente para **mala**.

Recordar la fórmula de Bashkara:$$x_{1,2}= \frac{-b\pm \sqrt{b^2 -4 \cdot a \cdot c}}{2 \cdot a}$$

**Tip**: podemos emplear la formula de despejar valores a partir de las raices, es decir: $$x_1 \cdot x_2 = \frac{c}{a}$$
Esto es a causa de que al momento de implementar la función, al restar dos números parecidos se pierden muchos datos dentro del algoritmo. Por lo tanto, calculamos una raíz (La de la suma) para posterior despejar la otra:
$$x_2 = \frac{c}{a\cdot x_1}$$

````python
def mala(a,b,c):
  '''
  PreCondición: Recibe 3 números de un polinomio de segundo grado
  PosCondición: Devuelve las raices mediante Bashkara
  '''
  x_1= 0
  x_2= 0
  x_1 = (-b+ math.sqrt(b**2 -4 * a *c))/2*a
  x_2 = (-b- math.sqrt(b**2 -4 * a *c))/2*a
  return(x_1,x_2)
````

````python
[x_1,x_2] = mala(1,2,-12)
print([x_1,x_2])
````

````python
def buena(a,b,c):
  '''
  PreCondición: Recibe 3 números de un polinomio de segundo grado
  PosCondición: Devuelve las raices mediante Bashkara
  '''
  x_1= 0
  x_2= 0
  x_1 = (-b+ ((b**2 -4 * a *c)**0.5))/2*a
  x_2 = c/(a*x_1)
  return(x_1,x_2)
````

**Ejercicio 9)** Escribir una función llamada *horn*, que implemente el algoritmo Horner para la evaluación de polinomios. La sintaxis de llamda a la rutina debe ser:
```
>>> p = horn(coefs,x)
```
donde `p` es el valor del polinomio, `coefs` es un vector con los coeficientes del polinomio, de mayor a menor grado y `x` es el valor de la variable independiente. Es decir que si, por ejemplo hacemos:
```
>>>p= horn([1,-5,6,2],2)
```
entonces la variable p almacenara el valor `p(2)` donde $p(x) = x^3-5x^2+6x+2$

````python
#Ejercicio 9)
def horn(coefs,x):
  '''
  PreCondicion: Recibe una lista, con coeficientes del polinomio, y un x con la variable independiente
  PostCondicion: Devuelve el resultado de valuar el polinomio
  '''
  n=0
  for i in range(len(coefs)):
    n = n * x + coefs[i]
  return n
````

````python
horn([1,-1],1)
````

**EJercicio 12.** Escribir una función llamda `Son Ortogonales([x1,x2],[y1,y2])` que tenga como entrada dos vectores $x,y \in \mathbb{R}n$ y que devuelve `True` si los vectores son ortogonales.

Ejecutar las sigiuentes instrucciones

```
x = [1,10240745112658109]
y = [-1, 1/x[1]]
if not SonOrtogonales(x,y):
  print("Algo salio mal..")
```

Explicar lo que sucedio

````python
#Ejercico 12
def SonOrtogonales(poli1, poli2):
  '''
  PreCondicion: Recibe dos listas con los valores correspondiente a los vectores
  Devuelve True si los valores son ortogonales
  '''
  suma = 0
  for i in range(len(poli1)):
    suma = suma + poli1[i]*poli2[i]
  if suma == 0:
    return True
````

````python
SonOrtogonales([1,2],[2,-1])
````

````python
x = [1,1.1024074512658109]
y = [-1,1/x[1]]
if not SonOrtogonales(x,y):
  print("Algo salio mal...")
````
