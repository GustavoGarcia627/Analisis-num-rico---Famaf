**Ejercicio 1** Programar una función en `python` qie integre numéricamente usando las reglas compuestas del trapecio, punto medio y Simpson. Nombrarla como `internumcomp`. La función deberá ejecutarse:



```
python3> S=internumcomp(fun,a,b,N,regla)
```
Donde

*   `fun`: Función de $\mathbb{R} \to \mathbb{R}$ a ser integrada
*   $a,b\in \mathbb{R}$ son los extremos de integración.
*   $N$ es la cantidad de subintervalos a usar
*   `regla` es un string que indica la regla que deberá ser `trapecio`, `pm` o `simpson`

La salida $S$ debe ser un número real





````python
def trapecio(fun,a,b,N):
  #Defino la longitud de subintervalos
  h = (b-a)/N

  #función evaluada en lo extremos
  fa = fun(a)
  fb = fun(b)
  sum = 0
  xi  = 0
  for i in range(1,N):
    xi = a+i*h
    sum+=fun(xi)
  res = (h/2) * (fa + 2*sum + fb)
  return res
````

````python
def pm (fun,a,b,N):
  #Defino longitud de subintervalos
  h = (b-a)/(N+2) #N debe ser par
  xi =0
  sum=0
  res=0
  n = int(N/2)
  for i in range(0,n+2,2):
    xi = a+((i+1)*h)
    sum+=fun(xi)
  res = 2*h*sum
  return res
````

````python
def simpson(fun, a, b, N):
    # Validación teórica: N tiene que ser par
    assert N % 2 == 0, "El número de subintervalos N debe ser par."

    h = (b - a) / N
    fa = fun(a)
    fb = fun(b)

    sx0 = fa + fb
    sx1 = 0 # Suma de f(x_{2j-1}) -> Nodos impares
    sx2 = 0 # Suma de f(x_{2j})   -> Nodos pares

    x = a
    for j in range(1, N):
        x = x + h
        if j % 2 == 0:  # j es par
            sx2 = sx2 + fun(x)
        else:           # j es impar
            sx1 = sx1 + fun(x)  # ¡Corregido de * a +!

    sx = (sx0 + 2 * sx2 + 4 * sx1) * (h / 3)
    return sx
````

````python
def internumcomp(fun,a,b,N,regla):
  s = 0
  if regla=="trapecio":
    s = trapecio(fun,a,b,N)
  elif regla=="pm":
    s = pm(fun,a,b,N)
  elif regla=="simpson":
    s = simpson(fun,a,b,N)
  else:
    return "regla incorrecta"
  return s
````

**Ejercicio 6** El período de un péndulo de longitud $l$ con amplitud $\alpha$ puede aproximarse con la fórmula:

$$
T= 4 \sqrt{\frac{l}{g}} \int_0^{\pi /2} \frac{d \theta}{\left(( 1-\sin^2 (\frac{\alpha}{2}) \sin^2(\theta) \right)^{\frac{1}{2}}}
$$

done $g = 9.8m/s^2$.

Programar una función, `pendulo`, que reciba una longitud $l$ en metros y $\alpha$ en forma de número entero entre 0 y 90, transforme el valor a radianes y devuelva el período del péndulo de longitud $l$. ¿Qué ocurre en el caso de $\alpha=0$?

````python
def pendulo(l: float, alpha: int):
    '''
    Pre: Recibe l en metros y alpha en [0,90]
    Pos: Devuelve el período del péndulo de longitud l
    '''
    g = 9.8 #constante de la gravedad

    # 1. Convertimos alpha a radianes
    alpha_rad = np.radians(alpha)

    # 2. Definimos la función a integrar
    def T(theta):
        denominador = (1 - (np.sin(alpha_rad / 2)**2) * (np.sin(theta)**2))**0.5
        return 1 / denominador

    # 3. Calculamos ambos factores de la ecuación
    factor_raiz = 4 * ((l / g) ** 0.5)
    integral, _ = integrate.quad(T, 0, np.pi / 2)

    # 4. Retornamos el período total
    return factor_raiz * integral
````

````python
print(pendulo(10,0))
````

Si se ejecuta la función $\alpha = 0$ entonces $\sin^2(0)=0$ donde el denominador queda como $\sqrt{1-0}=1$ donde el integrando resulta en $\int_0^{\pi/2}1 d\theta$ donde el resultado es simplemente $\frac{\pi}{2}$ dejando como resultado:

$$
T = 4 \sqrt{\frac{l}{g}} \cdot \frac{\pi}{2}= 2 \pi \sqrt{\frac{l}{g}}
$$

Resultando en la **fórmula clásica del período de un péndulo simple** para amplitudes pequeñas.

**Ejercicio 7:** Se desea implementar una regla de cuadratura adaptiva, es decir, una cuadratura compuesta que utilice más subintervalos en la zona en que la aproximación obtenida sea peor.

Para ello, notamos $S(a, b)$ a la regla de Simpson en el intervalo $[a, b]$. Si notamos $c = \frac{a+b}{2}$, se tiene que:

 $$\frac{|S(a,b)- S(a,c) - S(c,b)|}{15} \approx E(a,c,b)$$

 donde $E(a,c,b)$ es el error cometido al aplicar la regla compuesta: $S(a,b) + S(c,b)$. Implementar un programa que reciba como input una función $f$, un intervalo $[a,b]$ y una tolerancia que devuelve el valor $q_1 + q_2$. En caso contrario, se aplica el mismo criterio para integrar $f$ en los intervalos $[a,c]$ y $[c,b]$, con una tolerancia $\frac{\epsilon }{2}$.

 Probar el programa calculando: $$\int_0^1x e^{-x^2}dx = \frac{1}{2}(1-e^{-1}). $$

 Comparar los resultados (y los tiempos de ejecución) con los obtenidos por la regla de Simpson compuesta.

````python
import time
````

````python
def simpson_simple(f, a, b):
  '''
  pre: f: función continua, [a,b]: extremos del intervalo
  pos: Calcula la regla de Simpson simple en el intervalo [a, b]
  '''
  c = (a + b) / 2
  return ((b - a) / 6) * (f(a) + 4 * f(c) + f(b))
````

````python
def funEj7(f, I, tol):
  '''
  Pre: Recibe una función f, un intervalo I=[a,b] y una tolerancia tol
  Pos: Devuelve la integral de f sobre el intervalo usando cuadratura adaptativa
  '''
  a, b = I[0], I[1]
  c = (a + b) / 2

  # Aproximación gruesa (un solo intervalo)
  s_total = simpson_simple(f, a, b)

  # Aproximación fina (dos subintervalos)
  q1 = simpson_simple(f, a, c)
  q2 = simpson_simple(f, c, b)

  # Estimación del error según el enunciado
  error_estimado = abs(s_total - (q1 + q2)) / 15

  # Criterio de parada
  if error_estimado < tol:
    return q1 + q2
  else:
    # Recursión en las dos mitades con la mitad de la tolerancia
    integral_izquierda = funEj7(f, [a, c], tol / 2)
    integral_derecha = funEj7(f, [c, b], tol / 2)
    return integral_izquierda + integral_derecha
````

````python
f_ejemplo = lambda x: x * np.exp(-x**2)
I = [0, 1]
tol = 1e-6
valor_real = 0.5 * (1 - np.exp(-1))

print(f"Valor real de la integral: {valor_real}\n")

# 1. Medición de Cuadratura Adaptativa
t0 = time.perf_counter()
res_adaptativa = funEj7(f_ejemplo, I, tol)
t1 = time.perf_counter()
tiempo_adaptativa = t1 - t0

print("--- CUADRATURA ADAPTATIVA ---")
print(f"Resultado: {res_adaptativa}")
print(f"Error absoluto: {abs(res_adaptativa - valor_real)}")
print(f"Tiempo de ejecución: {tiempo_adaptativa:.6f} segundos\n")

# 2. Medición de Simpson Compuesta (ajustamos n para una precisión similar)
n_paneles = 100
t0 = time.perf_counter()
res_compuesta = internumcomp(f_ejemplo, I[0], I[1], n_paneles,"simpson")
t1 = time.perf_counter()
tiempo_compuesta = t1 - t0

print("--- SIMPSON COMPUESTA ---")
print(f"Resultado (n={n_paneles}): {res_compuesta}")
print(f"Error absoluto: {abs(res_compuesta - valor_real)}")
print(f"Tiempo de ejecución: {tiempo_compuesta:.6f} segundos")
````

Se observa que la cuadratura adaptativa es más efectiva en tiempos pero menos precisa
